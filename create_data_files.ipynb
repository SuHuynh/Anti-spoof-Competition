# Combine data for training

```python
import pandas as pd
import csv

train_path1 = '/data/CelebA/CelebA_Spoof_Face/train_high.csv'
train_path2 = '/data/CelebA/CelebA_Spoof_Face/train_low.csv'
train_path3 = '/data/CelebA/CelebA_Spoof_Face/train_middle.csv'

data1 = pd.read_csv(train_path1)
data2 = pd.read_csv(train_path2) 
data3 = pd.read_csv(train_path3) 
```

```python
a=pd.concat([data1,data2, data3], ignore_index=True)
```

```python
a.head()
```

```python
a.to_csv('data_train_all.csv', index=False)
```

# Combine data for testing

```python
import pandas as pd
import csv

test_path1 = '/data/CelebA/CelebA_Spoof_Face/test_high.csv'
test_path2 = '/data/CelebA/CelebA_Spoof_Face/test_low.csv'
test_path3 = '/data/CelebA/CelebA_Spoof_Face/test_middle.csv'

data_test_1 = pd.read_csv(test_path1)
data_test_2 = pd.read_csv(test_path2) 
data_test_3 = pd.read_csv(test_path3) 
```
